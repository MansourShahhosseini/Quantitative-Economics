##### **1.1 Data Generating Process**
###### **Write a Python code to fabricate the following process: $$\large y = 2 + 3x_1 - 5x_2 + \epsilon$$ where $x_1 \sim N(2,1)$, $x_2 \sim Binomial(5,0.3)$ and $\epsilon \sim N(0,2)$ and set sample size 300. To make sure that we all work with exactly the same data, set the random generator's seed number equal to 1403. Draw the histograms.** 

```python
def ols_generator(n, plot=True):
    np.random.seed(1403)
    x_1 = np.random.normal(2,1,n)
    x_2 = np.random.binomial(5,0.3,n)
    e = np.random.normal(0,np.sqrt(2),n)
    y = 2 + 3*x_1 - 5*x_2 + e
    df = pd.DataFrame({'x_1':x_1, 'x_2':x_2, 'y':y})
    if plot:
        sns.histplot(y, kde=True)
    return df

df = ols_generator(300)
```

##### **1.2 Estimation**
###### **Here we want to estimate $\beta$ using the following 4 methods:**
###### **1. Python method for OLS regression: Call it $\hat{\beta_{py}}$**
###### **2. Use algebra to find $\hat{\beta_{algebra}} = (XX')^{-1}X'Y$**
###### **3. Use Python optimizer to solve for $\hat{\beta_{optim}}$ hat minimizes the (SSR) sum of squared residual for the sample: $\sum (y_i - Xi\beta)^2$**
###### **4. Use Python optimizer to maximize the likelihood function in order to find $\hat{\beta_{ML}}$**
###### **Use sample size of n = 10, 20, 50, 100, 1000, 10000 to estimate $\beta$ and plot $[\hat{\beta_{py}}, \hat{\beta_{algebra}}, \hat{\beta_{optim}}, \hat{\beta_{ML}}]$ as a function of $n$. You should draw 3 plots for intercept, $\beta_1$ and $\beta_2$.**

```python
def old_python_lib(n):
    df = ols_generator(n, plot=False)
    X = sm.add_constant(df.drop('y', axis=1))
    Y = df.y
    result = sm.OLS(Y, X).fit()
    coefs = result.params.values
    confints = result.conf_int().values
    return coefs, confints
```

```python
ns = [10, 20, 50, 100, 1000, 10000]
all_coeefs = []
all_confints = []
for n in ns:
    all_coeefs.append(old_python_lib(n)[0])
    all_confints.append(old_python_lib(n)[1])

all_coeefs = np.array(all_coeefs)
all_confints = np.array(all_confints)

for i,c in enumerate(['darkblue', 'red', 'green']):
    sns.lineplot(x=range(len(ns)), y=all_coeefs[:,i], label = f'$\\hat{{\\beta_{i}}}$',  marker='o', color = c)
    sns.lineplot(x=range(len(ns)), y=all_confints[:,i,0], alpha = 0.1,  marker='o', color = c)
    sns.lineplot(x=range(len(ns)), y=all_confints[:,i,1], alpha = 0.1,  marker='o', color = c)
    plt.fill_between(range(len(ns)), all_confints[:,i,0], all_confints[:,i,1], alpha=0.2, color = c)
plt.xticks(np.arange(len(ns)), ns)
plt.yticks(np.arange(-6,6))
plt.xlabel("Number of Sample")
plt.grid()
plt.ylabel("$\\hat{\\beta}$ Estimation")
plt.title('OLS with Python Library')
plt.show()
```

```python
def ols_algebra(n):
    df = ols_generator(n, plot=False)
    df.insert(0, 'const', 1)
    X = df.drop('y', axis=1).values
    Y = df.y.values.reshape(-1,1)
    coefs = np.linalg.inv(X.T @ X) @ X.T @ Y
    return coefs
```

```python
ns = [10, 20, 50, 100, 1000, 10000]
all_coeefs = []
for n in ns:
    all_coeefs.append(ols_algebra(n))
all_coeefs = np.array(all_coeefs).squeeze()
for i,c in enumerate(['darkblue', 'red', 'green']):
    sns.lineplot(x=range(len(ns)), y=all_coeefs[:,i], label = f'$\\hat{{\\beta_{i}}}$',  marker='o', color = c)
plt.xticks(np.arange(len(ns)), ns)
plt.yticks(np.arange(-6,6))
plt.xlabel("Number of Sample")
plt.grid()
plt.ylabel("$\\hat{\\beta}$ Estimation")
plt.title('OLS with Algebra')
plt.show()
```

```python
def min_ssr(n):
    df = ols_generator(n, plot=False)
    df.insert(0, 'const', 1)
    betas = torch.zeros((3, 1), requires_grad=True)
    optimizer = torch.optim.Adam([betas], lr=0.001)
    X = torch.tensor(df.drop('y', axis=1).values, dtype=torch.float)
    Y = torch.tensor(df.y.values.reshape(-1,1), dtype=torch.float)
    num_iterations = 10000
    for i in range(num_iterations):
        optimizer.zero_grad()
        y_pred = X.mm(betas)
        loss = torch.sum((Y - y_pred) ** 2)
        loss.backward()
        optimizer.step()
        if (i + 1) % 2000 == 0:
            print(f"Iteration {i+1}, MSE: {loss.item():.4f}")
    return betas.detach().cpu().numpy()
```

```python
ns = [10, 20, 50, 100, 1000, 10000]
all_coeefs = []
for n in ns:
    all_coeefs.append(min_ssr(n))
all_coeefs = np.array(all_coeefs).squeeze()
for i,c in enumerate(['darkblue', 'red', 'green']):
    sns.lineplot(x=range(len(ns)), y=all_coeefs[:,i], label = f'$\\hat{{\\beta_{i}}}$',  marker='o', color = c)
plt.xticks(np.arange(len(ns)), ns)
plt.yticks(np.arange(-6,6))
plt.xlabel("Number of Sample")
plt.grid()
plt.ylabel("$\\hat{\\beta}$ Estimation")
plt.title('OLS with Min SSR')
plt.show()
```

```python
def max_loglike(n):
    df = ols_generator(n, plot=False)
    df.insert(0, 'const', 1)
    
    betas = torch.zeros((3, 1), requires_grad=True)
    sigma2 = torch.tensor([1.0], requires_grad=True)

    optimizer = torch.optim.Adam([betas,sigma2], lr=0.001)
    
    X = torch.tensor(df.drop('y', axis=1).values, dtype=torch.float)
    
    Y = torch.tensor(df.y.values.reshape(-1,1), dtype=torch.float)
    
    num_iterations = 10000
    
    for i in range(num_iterations):
        optimizer.zero_grad()
        y_pred = X.mm(betas)
        loss = -(-0.5 * n * torch.log(torch.tensor(2 * torch.pi)) - 0.5 * n * torch.log(sigma2) - (0.5 / sigma2) * torch.sum((Y - y_pred) ** 2))
        loss.backward()
        optimizer.step()
        if (i + 1) % 2000 == 0:
            print(f"Iteration {i+1}, MSE: {loss.item():.4f}")
    return betas.detach().cpu().numpy()
```

```python
ns = [10, 20, 50, 100, 1000, 10000]
all_coeefs = []
for n in ns:
    all_coeefs.append(max_loglike(n))
all_coeefs = np.array(all_coeefs).squeeze()
for i,c in enumerate(['darkblue', 'red', 'green']):
    sns.lineplot(x=range(len(ns)), y=all_coeefs[:,i], label = f'$\\hat{{\\beta_{i}}}$',  marker='o', color = c)
plt.xticks(np.arange(len(ns)), ns)
plt.yticks(np.arange(-6,6))
plt.xlabel("Number of Sample")
plt.grid()
plt.ylabel("$\\hat{\\beta}$ Estimation")
plt.title('OLS with Max Log Likelihood')
plt.show()
```

#### **2. A Search Model for Young Workers**
###### **As young workers enter labor market, they receive a job offer that they have to accept or reject. If they accept the job they will earn the offered wage and become employed. If they reject the oÂ¤er they will stay unemployed but enjoy their leisure time. Jobs are diÂ¤erent in two dimensions: hourly wage rate, and hours needed for the job. Employed workers remain employed forever and also unemployed workers remain unemployed forever. Unemployed workers receive an observable monetary subsidy of z as unemployment insurance. Then utility of workers can be modeled with $$\large U_{ni} = \alpha I_{ni} + \beta L_{ni} + \epsilon_{ni}$$ where $i = 1$ represent employment and $i = 0$. represent unemployment. Also income of employed workers are $\large I_{n1} = w_h h_h$ and unemployed workers'income is only the $UI: I_{n0} = z$. Their leisure are $\large L_{n1} = (1-h_h)$ and $\large L_{n0} = 1$ for employed and unemployed ones respectively. Of course, as researchers, we don't observe $U_{ni}$. What is observable is $y_{ni}$ which is a binary variable. Imagine that we also observe all job offers $(w_h,h_h)$ for everyone, either employed or unemployed. (This is not usually the case in practical studies. In the next problem set we will study the Roy model and selection problem that arises in empirical research.) Also notice that we assume that all workers have exactly the same taste toward income and leisure ($\alpha$ and $\beta$ are similar for all workers). We will later relax this assumption. Young individuals accept the wage offer if and only if $U_{n1} > U_{n0}$. Assume $\beta = 0$**


##### **2.1 Data Generating Process**
###### **Let's create such economy. Set random generator's seed number equal 1395 so we all get similar results. Assume $\epsilon_{ni}$ are iid and normally distributed: $\epsilon_{ni} \sim N(0, \sigma^2)$. Then use Python to simulate a sample of $1,000$ workers, once the true parameters of the model are $$\Theta \equiv (\alpha, \sigma) = (1, 3)$$ and wage oÂ¤ers are distributed exponentially with parameter $\lambda = 0.2: f(w) = 0.2 e^{-0.2w}$. Also $z = 12$ is observable by the researcher. Notice that here we assumed a very specific covariance matrix for error terms. We will relax this assumption later. Draw scatter plot of workers decisions versus their wage oÂ¤ers.** 

```python
def dgp_workers(n = 1000, alpha = 1, sigma = 3, w_lambda = 0.2, z = 12, plot=True):
    cp.random.seed(1395)
    e = cp.random.normal(0, sigma, (n,2))
    w = cp.random.exponential(scale = 1 / w_lambda, size = n)
    u_0 = alpha * z + e[:,0]
    u_1 = alpha * w + e[:,1]
    accept_offer = cp.where(u_1 > u_0, 1, 0)
    accept_offer, w = accept_offer.get(), w.get()
    if plot:
        sns.scatterplot(x = w, y = accept_offer, hue=np.where(accept_offer == 1, 'Accepted', 'Rejected'), palette={'Accepted': 'green', 'Rejected': 'red'})
        plt.xlabel('Wage')
        plt.ylabel('Choice')
        plt.title(f'Probabilty of Acceptance: {accept_offer.mean():.3f}')
        plt.show()
    return accept_offer, w
```

```python
observed_decision, observed_wage = dgp_workers()
```

##### **2.2 OLS, Logit and Probit**

```python
def model_fit(Y, X, kind, show_summary=False):
    if kind == 'logit':
        model = sm.Logit(Y.reshape(-1,1), sm.add_constant(X))
        result = model.fit()
    elif kind == 'probit':
        model = sm.Probit(Y.reshape(-1,1), sm.add_constant(X))
        result = model.fit()        
    elif kind == 'ols':
        model = sm.OLS(Y.reshape(-1,1), sm.add_constant(X))
        result = model.fit()
    if show_summary:
        print(result.summary())
    return result          
```

#### **3. Maximum Simulated Likelihood Estimator**

##### **3.1 Identification**
###### **Imagine that we are given data on $y_n$ and $x_n$. We assume that the true model is the one that we explained in the previous problem. In other words, we impose structure on the data to better understand the relation between observed variables. Benefit of this exercise is that we already know the true parameters. So we can start from a legitimate initial guess and we expect to find estimates close enough to the true parameters. Our goal is to estimate $\Theta = (\alpha, \Omega)$ where $\large \Omega = \begin{pmatrix} \sigma_{00} & \sigma_{01} \\ \cdot & \sigma_{11} \end{pmatrix}$. To take account of the fact that the level of utility is irrelevant, we take utility differences, therefore the only identified term is $\sigma^* = \sigma_{00} + \sigma_{11} - 2\sigma_{01}$. One possible identification assumption is to set $\sigma_{01} = 0$ and $\sigma_{00} = \sigma_{11}$ which may not be the best assumption depending on the context, but if we make this assumption then $\sigma_{00} = \sigma_{11} = \sigma^*/2$ will be identified. For the rest of this problem letâs impose this structure on the covariance matrix of error terms. Create a two-dimensional grid for $(\alpha, \sigma)$ which includes the true parameters: $(1, 3)$. For each of the points on this grid do the following:**
###### **1. Use the accept and reject simulator, discussed in class, to calculate $\tilde{P}_{ni}$ using $R = 100$ simulations.** 
###### **2. Then use the observed data, $(y_n, w_n)$, to calculate the simulated log likelihood. Recall that the log-likelihood function is $\mathcal{LL} = \sum_n \sum_j d_{nj} \log P_{nj}$, where $d_{nj} = 1$ if $n$ chose $j$ and 0 otherwise. When the probabilities cannot be calculated exactly, as in the case of Probit, the simulated log-likelihood function is used instead, with the true probabilities replaced with the simulated probabilities: $\mathcal{SLL} = \sum_n \sum_j d_{nj} \log \tilde{P}_{nj}$. Then find the parameters $(\hat{\alpha}, \hat{\sigma})$ that maximizes the above simulated likelihood function on this grid. If the maximum takes at a point close enough to $(1, 9)$, then congratulations! The value of the parameters that maximizes $\mathcal{SLL}$ is called the maximum simulated likelihood estimator (MSLE). You have successfully created your first MSLE!** 

```python
n = 1000
r = 100
```

```python
observed_decision, observed_wage = dgp_workers(plot=False, n=n)
```

```python
def simulate_workers(n = 1000, alpha = 1, sigma = 3, w_lambda = 0.2, z = 12, r = 100):
    cp.random.seed(1395)
    e = cp.random.normal(0, sigma, (r,n,2))
    w = cp.random.exponential(scale = 1 / w_lambda, size = (r,n))
    u_0 = alpha * z + e[:,:,0]
    u_1 = alpha * w + e[:,:,1]
    accept_offer = cp.where(u_1 > u_0, 1, 0)
    accept_offer_sim_ind = accept_offer.mean(0)
    accept_offer_sim_ind = accept_offer_sim_ind.get()
    return accept_offer_sim_ind
```

```python
def sll(alpha, sigma, n, r):
    accept_offer_sim_ind = simulate_workers(alpha=alpha, sigma=sigma, n=n, r=r)
    log_like = ((1 - cp.array(observed_decision)) * cp.log(1 - cp.array(accept_offer_sim_ind)) + cp.array(observed_decision) * cp.log(cp.array(accept_offer_sim_ind))).sum()
    return log_like.get()
```

```python
alphas = cp.arange(0.1,2.1,0.1)
sigmas = cp.arange(1,5.2,0.2)
grid = cp.zeros((len(alphas), len(sigmas)))

for a in range(len(alphas)):
    for s in range(len(sigmas)):
        grid[a,s] = sll(alpha=alphas[a], sigma=sigmas[s], n=n, r=r)

alphas[cp.where(grid == grid.max())[0][0]], sigmas[cp.where(grid == grid.max())[1][0]]   #(alpha_max, sigma_max)
```
